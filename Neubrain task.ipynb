```python
#updating the lowercase with uppercase in VEHICLE CLASS column
df["VEHICLE CLASS"].replace(to_replace=['Compact', 'SUV: Small',
       'Two-seater', 'Mid-size', 'Minicompact', 'Subcompact',
       'Station wagon: Small', 'Full-size', 'SUV: Standard',
       'Special purpose vehicle', 'Pickup truck: Small',
       'Pickup truck: Standard', 'Minivan', 'Van: Passenger',
       'Station wagon: Mid-size'],value=["COMPACT","SUV - SMALL","TWO-SEATER",
                                        "MID-SIZE","MINICOMPACT","SUBCOMPACT",
                                        "STATION WAGON - SMALL","FULL-SIZE","SUV - STANDARD",
                                        "SPECIAL PURPOSE VEHICLE","PICKUP TRUCK - SMALL","PICKUP TRUCK - STANDARD",
                                        "MINIVAN","VAN - PASSENGER","STATION WAGON - MID-SIZE"],inplace=True)
```

```python
#updating the lowercase words with uppercase in MAKE column
df['MAKE'].replace(to_replace=['Acura',
       'Alfa Romeo', 'Aston Martin', 'Audi', 'Bentley', 'Bugatti','Buick', 'Cadillac', 'Chevrolet', 'Chrysler', 
        'Dodge', 'Ford','Genesis', 'Honda', 'Hyundai', 'Infiniti', 'Jaguar', 'Jeep', 'Kia','Lamborghini', 'Land Rover', 'Lexus', 'Lincoln', 'Maserati',
        'Mazda', 'Mercedes-Benz', 'Mitsubishi', 'Nissan', 'Porsche', 'Ram','Rolls-Royce', 'Subaru', 'Toyota', 'Volkswagen', 'Volvo'],
        value=['ACURA','ALFA ROMEO','ASTON MARTIN','AUDI','BENTLEY','BUGATTI','BUICK','CADILLAC','CHEVROLET','CHRYSLER','DODGE','FORD','GENESIS',
             'HONDA','HYUNDAI','INFINITI','JAGUAR','JEEP','KIA','LAMBORGHINI','LAND ROVER','LEXUS',
            'LINCOLN','MASERATI','MAZDA','MERCEDES-BENZ','MITSUBISHI','NISSAN','PORSCHE','RAM',
            'ROLLS-ROYCE','SUBARU','TOYOTA','VOLKSWAGEN','VOLVO'],inplace=True)
```

```python
#replacing the values of Fuel column

df['FUEL'].replace(to_replace=['X', 'Z', 'E', 'N', 'D'],value=['Regular gasoline','Premium gasoline','Ethanol (E85)',
                                                               'Natural Gas','Diesel'],inplace=True)
```

```python
# Number of duplicated values contained in each column
df.duplicated().sum()
```

```python
# Drop duplicated rows
df.drop_duplicates(inplace=True)
```

# EXPLORATORY DATA ANALYSIS

```python
#Visualize the distribution of the fuel consumption: Using a histogram or box plot to visualize the distribution of fuel consumption in the dataset. 
#This will help to identify any outliers or unusual patterns in the data.
# Create a histogram of fuel consumption
plt.hist(df["FUEL CONSUMPTION"], bins=20)
plt.xlabel("Fuel Consumption")
plt.ylabel("Frequency")
plt.title("Distribution of Fuel Consumption")
plt.show()
```

```python
#This plot shows the distribution of the fuel consumption values in terms of their quartiles and outliers.
plt.boxplot(df["FUEL CONSUMPTION"])
plt.xlabel("Fuel Consumption")
plt.title("Box plot of Fuel Consumption")
plt.show()
```

```python
#To explore the relationship between engine size and fuel consumption, 
#we can create a scatter plot using the matplotlib library in Python.
plt.scatter(df['ENGINE SIZE'], df['FUEL CONSUMPTION'])
plt.xlabel('Engine Size')
plt.ylabel('Fuel Consumption (L/100 km)')
plt.title('Relationship between Engine Size and Fuel Consumption')
plt.show()
#as engine size increases fuel consumption also increases
```

```python
#Lets explore other relations also
sns.scatterplot(data=df, x='ENGINE SIZE', y='FUEL CONSUMPTION', hue='CYLINDERS')
```

```python
plt.xticks(rotation=90)
sns.scatterplot(data=df,x='FUEL',y='FUEL CONSUMPTION')
```

```python
plt.xticks(rotation=90)
sns.scatterplot(data=df, x='TRANSMISSION', y='FUEL CONSUMPTION', hue='CYLINDERS')
```

```python
sns.scatterplot(data=df,x='FUEL CONSUMPTION',y='EMISSIONS',hue='CYLINDERS')
#as seen from the plots there is a positive correlation between engine size,cylinder,and fuel consumption
```

```python
#To investigate the relationship between vehicle class and fuel consumption, we can create a bar chart to visualize 
#the average fuel consumption for each vehicle class in the dataset. This will help us understand 
#whether certain classes of vehicles tend to have higher or lower fuel consumption than others.
# group the data by vehicle class and calculate the mean fuel consumption for each group
mean_fuel_consumption = df.groupby('VEHICLE CLASS')['FUEL CONSUMPTION'].mean()

# create a bar chart of the mean fuel consumption for each vehicle class
plt.bar(mean_fuel_consumption.index, mean_fuel_consumption.values)
plt.xticks(rotation=90)
# add labels and title to the plot
plt.xlabel('Vehicle Class',fontsize = 2)
plt.ylabel('Average Fuel Consumption (L/100 km)')
plt.title('Average Fuel Consumption by Vehicle Class')

# display the plot
plt.show()

#As seen from the plot Station Wagon -small has lowest fuel consumption whereas VAN-Passenger has highest fuel consumption
```

```python
#AVERAGE EMISSIONS BY VEHICLE CLASS
mean_emission = df.groupby('VEHICLE CLASS')['EMISSIONS'].mean()

# create a bar chart of the mean fuel consumption for each vehicle class
plt.bar(mean_emission.index, mean_emission.values)
plt.xticks(rotation=90)
# add labels and title to the plot
plt.xlabel('Vehicle Class',fontsize = 2)
plt.ylabel('Emissions')
plt.title('Average Emissions by Vehicle Class')

# display the plot
plt.show()
```

```python
#Average engine size by transmission type
mean_fuel_consumption = df.groupby('TRANSMISSION')['ENGINE SIZE'].mean()

# create a bar chart of the mean fuel consumption for each vehicle class
plt.bar(mean_fuel_consumption.index, mean_fuel_consumption.values)
plt.xticks(rotation=90)
# add labels and title to the plot
plt.xlabel('Transmission',fontsize = 2)
plt.ylabel('Engine Size')
plt.title('Average Transmission by Engine Size')

# display the plot
plt.show()
```

```python

# Group the data by year and calculate the mean fuel consumption
fuel_by_year = df.groupby('YEAR')['FUEL CONSUMPTION'].mean()

# Create a bar chart for the average fuel consumption by year
plt.bar(fuel_by_year.index, fuel_by_year.values)

# Set the title and axis labels
plt.title('Average Fuel Consumption by Year')
plt.xlabel('Year')
plt.ylabel('Fuel Consumption (L/100 km)')

# Show the plot
plt.show()
#mean fuel consumption for all the years from 2000 to 2022 is between 12 to 14
```

```python
#pie_chart of the top 10 highest polluting vehicles
top_10_highest_emission_vehicles=df.groupby(by=['MAKE','VEHICLE CLASS','FUEL']).sum()['EMISSIONS'].nlargest(10).reset_index()
fig,ax = plt.subplots(figsize=(10,10))
plt.pie(x='EMISSIONS',labels='MAKE',data=top_10_highest_emission_vehicles)
```
